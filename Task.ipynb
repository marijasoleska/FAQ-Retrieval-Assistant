```python
import numpy as np
import pandas as pd
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('mock_database.csv')
```

```python
dataframe = pd.DataFrame(data)
```

```python
dataframe.drop(columns=["id", "category"], inplace=True)
```

```python
model = SentenceTransformer('paraphrase-multilingual-MiniLM-L12-v2')
```

```python
questions = dataframe['question'].tolist()
```

```python
q_embeddings = model.encode(questions, show_progress_bar=True,normalize_embeddings=True)
```

```python
#TEST1 QUERY
q = "I forgot my password, how can I recover it?"
```

```python
q_embedding = model.encode([q],normalize_embeddings=True)[0]
```

```python
print(q_embedding[:11])
```

```python
def find_top_3(q): 
    if not q:
        print("Please enter a valid question.")
    query_embedding = model.encode([q],normalize_embeddings=True)
    similarities = q_embeddings @ query_embedding[0]
    top_3 = np.argsort(similarities)[-3:][::-1]
    
    results = []
    for index in top_3:
        results.append({
            "Question": dataframe.iloc[index]["question"],
            "Answer": dataframe.iloc[index]["answer"],
            "Confidence": similarities[index]
        })

    return results
```

```python
#TEST 2 QUERY
query = input("Enter your question (EN/MK/DE): ")
results = find_top_3(query)

print("\nTop 3 relevant FAQs:\n")

for r in results:
    print(f"Question: {r['Question']}")
    print(f"Answer: {r['Answer']}")
    print(f"Confidence: {r['Confidence']:.2f}\n")

print("Best matching answer:")
print(results[0]["Answer"])
print(f"Confidence: {results[0]['Confidence']:.2f}")
if results[0]["Confidence"] < 0.45:
    print("Low confidence")
```

```python
for q in [
    "I forgot my password",
    "Не можам да се најавам, ја заборавив лозинката",
    "I was charged twice on my card",
]:
    r = find_top_3(q)
    print("\nQuery:", q)
    print("Best answer:", r[0]["Answer"])
    print("Confidence:", round(r[0]["Confidence"], 2))
```
